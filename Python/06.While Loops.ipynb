```python
number = 0

while number < 5:
    print(number)
    number = number + 1
```

```python
#break
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
# else statement 
number = 0

while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('No Longer < 5')
```

```python
# continuous statement
number = 0

while number < 5:
    number = number + 1
    
    if number == 3:
        continue
    print(number)
else:
    print('No Longer < 5')
```

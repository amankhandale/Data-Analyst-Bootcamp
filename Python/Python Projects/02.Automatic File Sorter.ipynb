Automatic File Sorter

```python
import os, shutil
```

```python
path = r"C:/Users/amank/OneDrive/Documents/Python Project/"
```

```python
file_name = os.listdir(path)
```

```python
os.path.exists(path + 'csv files')
```

```python
folder_names = ['csv files', 'image files']

for loop in range(0,2):
    if not os.path.exists(path + folder_names[loop]):
        print(path + folder_names[loop])
        os.makedirs(path + folder_names[loop])
```

```python
for file in file_name:
    if ".csv" in file and not os.path.exists(path + "csv files/" + file):
        shutil.move(path + file, path + "csv files/" + file)
    elif ".png" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    else:
        print("No Changes Maybe The File have already been Moved")
```

This will be useful if you set timer for the execution for every 5min or 30mins (That will automatically sort incoming file in the folder)

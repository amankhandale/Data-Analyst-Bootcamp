```python
def number_squared(number):
    print(number**2)  #this will give the number power of 2
```

```python
number_squared(5)
```

```python
def number_squared_cust(number,power):
    print(number**power)  #this will give you the option to set custom power value
```

```python
number_squared_cust(5,5)
```

```python
# Arbitary Augments
def number_args(*number):
    print(number[0]*number[1])
```

```python
number_args(5,6,1,2,8) # The above argument has [0], [1]

#which means 5 * 6 = 30 as mentioned in the function command
```

```python
args_tuple = (5,6,1,2,8)

def number_args(*number):
    print(number[0]*number[1])
```

```python
number_args(*args_tuple)
```

```python
number_squared_cust(power = 5,number = 3) #keyword argument
```

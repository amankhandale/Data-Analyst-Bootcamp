Logical Operators

```python
(10 > 50) and (50 > 10) # (AND) = Returns True only when the both statements are true
```

```python
(70 > 50) and (50 > 10) # AND True statement = Both statement True
```

```python
(10 > 50) or (50 > 10) # (OR) = Returns True if one statement is True
```

```python
(10 > 50) or (50 < 10) # OR False statement = Both statement is False
```

```python
('Vanilla' > 'Chocolate') or (50 > 10) # 'V' is Greater than C = True
```

```python
not(50 > 10) # Returns True when the statement is actually false (Inverted Result)
```

Membership Operators

```python
ice_cream = 'I love Chocolate ice cream'
'Chocolate' in ice_cream 

#'in' Checks the string (if 'Chocolate' is included) = True
```

```python
scoops = [1,2,3,4,5]

6 in scoops # if '6' doesn't exist int variable then = False
```

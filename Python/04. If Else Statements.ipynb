If - Elif - Else Statements

```python
if 25 > 10:
    print('It Worked!') # this statement will only execute when 'if' is true
```

```python
if 25 < 10:
    print('It Worked!') # Here the statement is False (so it won't execute the command below)
```

```python
if 25 < 10:
    print('It Worked!') # False Statement = no execution
else:
    print("Nope, Didn't work") # Instead this will execute when the if statement is False
```

```python
if (25 < 10) or (1 < 3):
    print('It worked!')
    if 10 > 5:
        print('This nested if statement worked!')
elif 25 < 20:
    print ('elif worked!')
elif 25 < 21:
    print ('elif 2 worked!')
elif 25 > 20:
    print('elif 3 worked!') # Elif can be used to check multiple statement in between if and else
elif 25 > 28:
    print('elif 4 worked!')
else:
    print("Didn't Work")
```

```python
print('it worked!') if 10 > 30 else print ('it didnt work')
```

Converting Data Types

```python
num_int = 7
type(num_int)
```

```python
num_str = '7'
type(num_str)
```

Converting String to Integer

```python
num_str_conv = int(num_str) #converting str var -> int var 
type(num_str_conv)
```

```python
num_sum = num_int + num_str_conv
print(num_sum)
```

Converting Lists , Sets and Tuples

```python
list_type = [1,2,3]
type(list_type)
```

```python
type(tuple(list_type)) # list -> Tuple
```

```python
list_type = [1,2,3,3,2,1,2,3,2,1]
```

```python
type (set(list_type))  #list -> Set
print(set (list_type))
```

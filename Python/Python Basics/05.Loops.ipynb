```python
ice_cream_dict = {'name': 'Aman', 'weekly intake': 5, 'favourite': ['Chocolate','Vanilla']}
```

```python
for cream in ice_cream_dict.values():
    print(cream)
```

```python
for key, value in ice_cream_dict.items():
    print(key, '->' , value)
```

Nested For Loops

```python
flavours = ['Vanilla','Chocolate', 'Blueberry']
toppings = ['Oreos', 'Cookies', 'Marshmallows']
```

```python
for one in flavours:
    for two in toppings:
        print(one, 'topped with', two)
```

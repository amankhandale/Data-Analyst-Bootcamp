```python
#Sets  
# it doesn't count duplicate values
daily_pints = {1,2,3}
```

```python
type(daily_pints)
```

```python
daily_pints_log = {1,2,3,2,1,4,5}
print(daily_pints_log)
```

```python
wifes_daily_pints_log = {1,3,5,7,3,24,6,8,3,23,0}
```

```python
print(daily_pints_log | wifes_daily_pints_log)
```

```python
print(daily_pints_log - wifes_daily_pints_log)
```

```python
print(wifes_daily_pints_log ^ daily_pints_log) # unique values
```
